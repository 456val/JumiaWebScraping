<h3><b>Scraping Data with BeautifulSoup</b></h3>

```python

import requests
from bs4 import BeautifulSoup


base_url="https://www.jumia.com.ng/android-phones"
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0"
}
end_page = 50
product_details = []
current_price = []
discount_price = []
links= []
star_rating = []
stock = []
number_of_reviews = []
actual_reviews = []
product_model=[]
product_weight=[]
product_color= []



for end in range(1,end_page+1):
    url = f'{base_url}?page={end}'
    response = requests.get(url, headers=header)
    soup= BeautifulSoup(response.content, 'html.parser')
    
    # Loop through each product on the page
    for product in soup.find_all("article", class_="prd"):
        # Extract Product Name and Link
        name_tag = product.find("h3", class_="name")
        product_name = name_tag.text.strip() if name_tag else "N/A"

        link_tag = product.find("a", href=True)
        product_link = "https://www.jumia.com.ng" + link_tag['href'] if link_tag else None

       

        # Extract Current Price
        price_tag = product.find("div", class_="prc")
        product_price = price_tag.text.strip() if price_tag else "N/A"

        # Extract Discount Price
        old_price_tag = product.find("div", class_="old")
        product_discount_price = old_price_tag.text.strip() if old_price_tag else "N/A"

        # Extract Stock Availability
        stock_tag = product.find("p", class_="-df -i-ctr -fs12 -phm -ptxs -rd5")
        product_stock = stock_tag.text.strip() if stock_tag else "In Stock or Unknown"
        
        # Extract Star Rating
        rating_tag = product.find("div", class_="stars _s")
        product_rating = rating_tag.text.strip() if rating_tag else "No rating"
        
       
        # Extract Number of Reviews
        reviews_tag = product.find("p", class_="-fs12 -pvxs")
        product_reviews = reviews_tag.text.strip() if reviews_tag else "0"
        
        reviews_text = product.find("p", class_="-pvxs")
        product_text = reviews_text.text.strip() if reviews_text else "0"
        
        

        # Extract Features (Model, Weight, Color)
        features_tag = product.find("ul", class_="-pvs -mvxs -phm -lsn")  # Class to match the <ul> that contains features

        model = weight = color = "N/A"  # Default values for Model, Weight, Color

        if features_tag:
            for feature in features_tag.find_all("li", class_="-pvxs"):
                span_tag = feature.find("span", class_="-b")  # Extract the text that identifies the feature (e.g., "Model", "Weight")
                if span_tag:
                    feature_label = span_tag.text.strip()

                    # Check for Model, Weight, and Color
                    if "Model" in feature_label:
                        model = feature.text.split(":")[-1].strip()
                    elif "Weight" in feature_label:
                        weight = feature.text.split(":")[-1].strip()
                    elif "Color" in feature_label:
                        color = feature.text.split(":")[-1].strip()

        # Append the data to the respective lists
        product_details.append(product_name)
        links.append(product_link)
       
        current_price.append(product_price)
        discount_price.append(product_discount_price)
        stock.append(product_stock)
        star_rating.append(product_rating)
        number_of_reviews.append(product_reviews)
        actual_reviews.append(product_text)
        product_model.append(model)  # Append extracted model
        product_weight.append(weight)  # Append extracted weight
        product_color.append(color)  # Append extracted color

    #print(f"Scraped page {end}")

# Optional: Print the first 5 results to verify
for i in range(5):
    print(f"Product {i+1}:")
    print(f"Name: {product_details[i]}")
    print(f"Link: {links[i]}")
    print(f"Price: {current_price[i]}")
    print(f"Discount Price: {discount_price[i]}")
    print(f"Stock: {stock[i]}")
    print(f"Rating: {star_rating[i]}")
    print(f"Reviews: {actual_reviews[i]}")
    print(f"Model: {product_model[i]}")
    print(f"Weight: {product_weight[i]}")
    print(f"Color: {product_color[i]}")
    print('-' * 40)
    
    
    
    
    
```

<h3><b>Load Data in DataFrame</b></h3>

```python
import pandas as pd

# Create a DataFrame from your lists
data = {
    "Product Name": product_details,
    "Link": links,
    "Current Price": current_price,
    "Discount Price": discount_price,
    "Stock": stock,
    "Rating": star_rating,
    "Number of Reviews": number_of_reviews,
    "Actual Review Text": actual_reviews,
    "Model": product_model,
    "Weight": product_weight,
    "Color": product_color
}

df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("jumia_android_phones.csv", index=False, encoding='utf-8-sig')

print("✅ Data saved to jumia_android_phones.csv")
```
